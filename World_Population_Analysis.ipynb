```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)'
```

```python
data = pandas.read_html(url)
```

```python
data[0].head()
```

```python
data[0].drop(0, axis = 0, inplace = True)
```

```python
data[0]['Country or territory'] = data[0]['Country or territory'].replace(to_replace = 'China[a]', value = 'China')
```

```python
df = data[0].sort_values(by = 'Population (1 July 2023)', ascending = False).head(5)
```

```python
df
```

```python
df.plot(kind = 'pie', y = 'Population (1 July 2023)', labels = df['Country or territory'], startangle = 90)
plt.title('Top Populated Countries 2023')
```

```python
df = data[0].groupby(by = 'UN continental region[1]')['Population (1 July 2023)'].sum()
```

```python
df.plot(kind = 'pie', labels = df.index)
plt.title('Top Populated Regions 2023')
```

```python
data[0].dtypes
```

```python
df = data[0].sort_values(by = 'Population (1 July 2022)', ascending = False).head(5)
```

```python
df.plot(kind = 'bar', y = 'Population (1 July 2022)', x = 'Country or territory', color = 'r')
plt.title('Top Populated Countries 2022')
```

```python
df = data[0].groupby(by = 'UN continental region[1]')['Population (1 July 2022)'].sum()
```

```python
df
```

```python
df.plot(kind = 'bar')
plt.title('Top Populated Regions 2022')
plt.xlabel('Regions')
```

```python
df = data[0].sort_values(by = 'Population (1 July 2023)', ascending = True).head()
```

```python
df.plot(kind = 'bar', x = 'Country or territory', y = 'Population (1 July 2023)', color = 'y')
plt.title('Least Populated Countries 2023')
```
