```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_Cup_and_UEFA_Champions_League_finals'
```

```python
data = pandas.read_html(url)
```

```python
data = data[1]
```

```python
data.head()
```

```python
print('Last Champion')
data.tail(1)
```

```python
print('First Champion')
data.head(1)
```

```python
data['Winners'].value_counts()
```

```python
winners = data['Winners'].value_counts().head(8)
```

```python
winners.plot(kind = 'bar')
plt.title('Top Winners')
```

```python
winner_country = data['Country'].value_counts()
winner_country
```

```python
winner_country = winner_country.head(5)
```

```python
winner_country.plot(kind = 'bar', color = 'r')
plt.title('Top Winner Countries')
```

```python
runner_up = data['Runners-up'].value_counts()
runner_up
```

```python
runner_up = runner_up.head(5)
```

```python
runner_up.plot(kind = 'pie')
plt.title('Teams losing most finals')
```

```python
runner_up_country = data['Country.1'].value_counts()
runner_up_country
```

```python
runner_up_country = runner_up_country.head(5)
```

```python
runner_up_country.plot(kind = 'pie')
plt.title('Countries losing most finals')
```
